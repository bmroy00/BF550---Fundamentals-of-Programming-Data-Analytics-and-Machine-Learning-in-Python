```python
import numpy as np
```

```python
# Define a 2*4 matrix "X" with 2 samples and 4 variables
X=np.zeros([2,4])
X=np.array([[1,2,3,4],[5,9,4,6]])
```

```python
# Problem 1 Part 1

# Use np.mean to take the mean of each variable

def matrix_mean(X):
    # Get rows and cols for nested for loops
    num_cols = len(X[0])
    num_rows = len(X)
    
    # Initialize a new matrix that will hold out mean values
    M = np.zeros([1,num_cols])
    
    # Add column values together across rows and divided them by number of rows
    for i in range(num_cols):
        sumX = 0
        for j in range(num_rows):
            sumX += X[j][i]
            
        M[0][i] = sumX / len(X)
        
    return(M)   

matrix_mean(X)
```

```python
# Problem 1 Part 2

# Take the covariance of the matrix

def matrix_cov(X):
    # Take the mean of matrix with function we just wrote
    X_mean = matrix_mean(X)
    
    # Subtract the mean from each value
    X_center = X - X_mean
    
    # Multiply the values we just created by itself
    cov = np.matmul(X_center.T,X_center)/len(X)
    return(cov)
    
matrix_cov(X)
```
